```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('0000000000002429_training_titanic_x_y_train.csv')
df
```

```python
df['Age'] = df['Age'].fillna(24)
```

```python
df['Sex'] = pd.get_dummies(df['Sex'],drop_first=True)
```

```python
x = df.iloc[:,:-1]
x
```

```python
df.drop(['Cabin','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
df
```

```python
x_train = df.iloc[:,:-1]
x_train
```

```python
y_train = df.iloc[:,-1]
y_train
```

```python
clf = LogisticRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train)
```

```python
dft = pd.read_csv('0000000000002429_test_titanic_x_test.csv')
dft
```

```python
dft['Age'] = dft['Age'].fillna(24)
```

```python
dft['Sex'] = pd.get_dummies(dft['Sex'],drop_first=True)
```

```python
dft.drop(['Cabin','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
dft
```

```python
x_test = dft
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
np.savetxt('titanic_prediction.csv',y_pred, delimiter=',', fmt='%1.5f')
```

```python
clf.score(x_test,y_pred)
```
